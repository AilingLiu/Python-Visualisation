```
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

This notebook follows the tutorial by 'New Planet School' on [Youtube](https://www.youtube.com/watch?v=in8LJ-aR8T8&list=PLwQ7o9tyw625ThzACVnUoO0XIMorcP52-&index=2).

## exponential decay

```
x_fake = np.linspace(0, 4, 20)
noise = 0.025*np.random.normal(size=len(x_fake))
y_fake = np.exp(-x_fake) + noise

x_theory = np.linspace(0, 5, 200)
y_theory = np.exp(-x_theory)
```

```
fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(x_theory, y_theory, 'g', label='theory', lw=3)
ax.plot(x_fake, y_fake, 'bo', label='data')
ax.set(xlabel='x', ylabel='data and theory', title = 'results from the experiment')
ax.legend()
ax.grid();
```

## log linear transformation of exponential decay

```
fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(x_theory, y_theory, 'g', label='theory', lw=3)
ax.plot(x_fake, y_fake, 'bo', label='data')
ax.set(xlabel='x', ylabel='data and theory', title = 'results from the experiment')
ax.legend()
ax.grid()
ax.set_yscale('log');
```

## error bars

```
x = np.linspace(0.1, 0.8, 8)
y = [.9, .8, .366, .5, .39, .45, .3, .15]
y_err = [.12, .05, .15, .17, .07, .2, .1, .14]

xx = np.linspace(0, 1, 50)
yy = np.exp(-3*xx/2)

fig, ax = plt.subplots(figsize=(6, 6))
ax.errorbar(x, y, yerr=y_err, lw=2, ecolor='g', fmt='o--', mew=2, label='data')

ax.plot(xx, yy, lw=3, color='magenta', label='theory')

ax.set(title='bacetria slowly die out', xlabel='time(hours)', 
       ylabel='number of bacgeria(normalized)')
ax.legend()
ax.grid();
```

## bar graphs

```
years = np.arange(1999, 2013, 1)
np.random.seed(42)
Denver = np.random.randint(1234, 9000, size=len(years))
Philly = np.random.randint(1106, 5503, len(years))

fig, ax = plt.subplots(figsize=(8, 6))
bwidth = 0.4
x = np.arange(len(years))

ax.bar(x, Denver, width=bwidth, color='green', label='Denver', alpha=0.5)
ax.bar(x+bwidth, Philly, width=bwidth, color='orange', label='Philly', alpha=0.5)
ax.set(xlabel='year', ylabel='Number of pennies in circulation',
       title='Pennies from Denver and Philadelphia Mints')
ax.set_xticks(x+(bwidth/2))
ax.set_xticklabels(years)
ax.set_yticks(np.arange(0, 10000, 1000))
ax.legend();
```

## stacked bar graph

```
years = ['y'+str(i) for i in range(0, 14)]
fig, ax = plt.subplots(figsize=(8, 6))

x = np.arange(0, 14, 1)
bwidth=0.4

ax.bar(x, Denver, width=bwidth, color='green', label='Denver')
ax.bar(x, Philly, width=bwidth, bottom=Denver, color='orange', label='Philadelphia')
ax.set(xlabel='year', ylabel='nuber of pennis in circulation', title='Pennies from Denver and Philadelphia Mints')
ax.set_xticks(x)
ax.set_xticklabels(years)
ax.legend();
```

## colored backgrounds and grids

```
x = np.linspace(0, 2*np.pi, 80)
y = np.cos(2.5*x)*np.exp(-x/2)

#as the number of points increase, the yellow line is closely approximating the original line
xc5=np.linspace(0, 2*np.pi, 5)
xc20=np.linspace(0, 2*np.pi, 20)
yc5=np.cos(2.5*xc5)*np.exp(-xc5/2)
yc20=np.cos(2.5*xc20)*np.exp(-xc20/2)

fig, (ax1, ax2)=plt.subplots(1, 2, figsize=(10, 6))

ax1.set_facecolor('slategrey')
ax1.plot(x, y, color='blue', lw=2)
ax1.plot(xc5, yc5, color='y', ls='--' ,marker='o', lw=3, label='Num of Obs: 5')
ax1.set(xlabel='time', ylabel='decaying signal(volts)', 
       title='calculus: approximating a function by a series of lines')
ax1.grid(color='white')
ax1.legend()

ax2.set_facecolor('slategrey')
ax2.plot(x, y, color='blue', lw=2)
ax2.plot(xc20, yc20, color='y', ls='--' ,marker='o', lw=3, label='Num of Obs: 20')

ax2.set(xlabel='time', ylabel='decaying signal(volts)', 
       title='calculus: approximating a function by a series of lines')

ax2.grid(color='white')
ax2.legend()

plt.tight_layout();
```

## adding text and arrow to the plot

```
fig, ax=plt.subplots(figsize=(6, 4))

ax.set_facecolor('slategrey')
ax.plot(x, y, color='blue', lw=2)
ax.plot(xc5, yc5, color='y', ls='--' ,marker='o', lw=3, label='Num of Obs: 5')
ax.set(xlabel='time', ylabel='decaying signal(volts)', 
       title='calculus: approximating a function by a series of lines')

ax.text(2, 0.75, 'this is the seconda point\n where gradient d=0', 
        fontsize=10,
        color='white')
ax.arrow(2, 0.65, 0.20, -0.15, shape='full', lw=2, width=0.05)

ax.tick_params(axis='both', which='major', labelsize=15)

ax.grid(color='white')
ax.legend();
```

## more than one plot together

```
t = np.linspace(0, 8, 80)
y_ud = np.exp(-t)*np.cos(2*t)
y_cd = (1+t)*np.exp(-t)
y_od = np.exp(-t)

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 9))

ax1.plot(t, y_ud, color='magenta', label='under demand')
ax1.grid()
ax1.legend()


ax2.plot(t, y_cd, color='red', label='critically demand')
ax2.set_ylabel('amplitude(meters)', fontsize=14)
ax2.set_yticks(np.arange(0, 1.2, 0.2))
ax2.tick_params('y')
ax2.grid()
ax2.legend()

ax3.plot(t, y_od, color='orange', label='over damped')
ax3.set_xlabel('time(seconds)', fontsize=14)
ax3.tick_params('x')
ax3.grid()
ax3.legend();
```
